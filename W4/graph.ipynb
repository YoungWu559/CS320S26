```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(comment='Balanced Binary Tree')
dot.attr(rankdir='TB')

# Add nodes
dot.node('1', '1')
dot.node('2', '2')
dot.node('3', '3')
dot.node('4', '4')
dot.node('5', '5')
dot.node('6', '6')
dot.node('7', '7')

# Add edges to create a balanced binary tree with height 2
dot.edge('1', '2')
dot.edge('1', '3')
dot.edge('2', '4')
dot.edge('2', '5')
dot.edge('3', '6')
dot.edge('3', '7')
dot.edge('4', '7')

# Display the tree
#dot.render('balanced_binary_tree', view=True)
#dot.view()
dot
```
