```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
    def tot(self):
        if self.next == None:
                return self.val
        return self.val + self.next.tot()
    def __getitem__(self, idx):
        if idx == 0:
            return self.val
        return self.next[idx-1]
A = Node(3)
B = Node(5)
C = Node(7)
A.next = B
B.next = C
```

```python
print(C.tot(), B.tot(), A.tot())
print(A[0], A[2])
print(A[-1])
C.next = A
print(C[3])
print(A.tot())
```
