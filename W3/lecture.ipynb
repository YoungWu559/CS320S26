```python
import zipfile
import csv
import io
```

```python
z = zipfile.ZipFile("wi.zip");
f = z.open("wi.csv")
c = csv.DictReader(io.TextIOWrapper(f, encoding='utf-8'))
sum = 0
count = 0
for row in c:
    # not c[row]
    try:
        sum += float(row["interest_rate"])
        count += 1
    except ValueError:
        sum += 0
        count += 1
sum / count
```

```python
def gen_rate():
    z = zipfile.ZipFile("wi.zip");
    f = z.open("wi.csv")
    c = csv.DictReader(io.TextIOWrapper(f, encoding='utf-8'))
    for row in c:
        try:
            yield float(row["interest_rate"])
        except ValueError:
            yield 0

g = gen_rate()
print(next(g))
print(next(g)) # if next is called, then for r in g will not restart from the beginning!
sum = 0
count = 0
for r in g: # This for loop starts from the third element
    sum += r
    count += 1
sum / count
# 2.2645360815930506
```

```python
# How to do this (read one line at a time) to compute the median???
g = gen_rate() # restart the list!!!
l = list(g)
l.sort()
median = l[len(l) // 2] if len(l) % 2 == 1 else (l[len(l) // 2 - 1] + l[len(l) // 2]) / 2
median
```
