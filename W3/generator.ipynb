```python
import zipfile
import csv
import io
```

```python
z = zipfile.ZipFile("wi.zip");
f = z.open("wi.csv");
total = 0
count = 0
r = csv.DictReader(io.TextIOWrapper(f, encoding = "utf-8"))
for row in r:
    try:
        total += float(row["interest_rate"])
        count += 1
    except ValueError:
        pass
f.close()
z.close()
total / count
```

```python
def gen_rates():
    z = zipfile.ZipFile("wi.zip")
    f = z.open("wi.csv")
    r = csv.DictReader(io.TextIOWrapper(f, encoding = "utf-8"))
    for row in r:
        try:
            yield float(row["interest_rate"])
        except ValueError:
            pass
    f.close()
    z.close()
rates = gen_rates()
print(next(rates), next(rates), next(rates))
total = 0
count = 0
for rate in rates:
    total += rate
    count += 1
total / count
```
