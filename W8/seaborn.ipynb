```python
import matplotlib.pyplot as plt
import seaborn
import pandas
```

```python
grades = pandas.read_csv("Grades.csv")
data = grades[(grades["Project"] != 0) & (grades["Exam"] != 0)]
seaborn.relplot(data, x = "Quiz", y = "Exam", size = "Exam", col = "Section", row = "Lab")
```

```python
seaborn.relplot(data, x = "Quiz", y = "Exam", size = "Project", hue = "Lab", col = "Section")
```

```python
copy = data.copy()
copy["Lab"] = data["Lab"].astype("category")
seaborn.relplot(copy, x = "Quiz", y = "Exam", size = "Project", hue = "Lab", col = "Section")
```

```python
seaborn.pairplot(copy.iloc[:, 0:4])
```
