```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
import pandas
import matplotlib.pyplot as plt
import itertools
```

```python
service = Service(executable_path="chromedriver-win64/chromedriver.exe")
driver = webdriver.Chrome(service=service)
url = "https://pages.cs.wisc.edu/~yw/CS320W26W6.html"
driver.get(url)
tables = driver.find_elements("tag name", "table")
table = tables[1]
rows = table.find_elements("tag name", "tr")
array = []
for row in rows:
    cols = row.find_elements("tag name", "td")
    lst = []
    for col in cols:
        lst.append(col.text)
    array.append(lst)
pandas.DataFrame(array)
```

```python
driver.quit()
```

```python
service = Service(executable_path="chromedriver-win64/chromedriver.exe")
driver = webdriver.Chrome(service=service)
url = "https://pages.cs.wisc.edu/~yw/CS320W26W6.html"
driver.get(url)
checkbox = [driver.find_element("id", f"answer{i}_mc") for i in range(1, 7)]
button = driver.find_element("id", "button_5")
field = driver.find_element("id", "out_mc")
answers = ["A", "B", "C", "D", "E", "F"]
```

```python
for choice1 in range(6):
    for choice2 in range(6):
        if choice1 != choice2:
            checkbox[choice1].click()
            checkbox[choice2].click()
            button.click()
            time.sleep(0.01)
            print(answers[choice1] + "," + answers[choice2] + ": " + field.text)
            checkbox[choice1].click()
            checkbox[choice2].click()
```

```python
driver.quit()
```

```python
service = Service(executable_path="chromedriver-win64/chromedriver.exe")
driver = webdriver.Chrome(service=service)
url = "https://pages.cs.wisc.edu/~yw/CS320W26W6.html"
driver.get(url)
input = driver.find_element("id", "answer_ps")
button = driver.find_element("id", "button_6")
output = driver.find_element("id", "out_ps")
```

```python
passwords = list(itertools.permutations(range(0, 10), 4)) # This takes too long
start = ""
passwords = list(itertools.permutations(range(0, 10), 2)) # Guess the last two digits
start = "84"
for password in passwords:
    keys =  start + "".join(list(map(str, password)))
    input.send_keys(keys)
    button.click()
    time.sleep(0.01)
    if not output.text.startswith("I"):
        print(keys + ": " + output.text)
    input.clear()
```

```python
driver.quit()
```
